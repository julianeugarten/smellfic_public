```python
import pandas as pd
```

```python
df = pd.read_csv('MythFic-frames_redacted.tsv', sep='\t', header=0, encoding='utf-8')
```

# A first look at some most-frequent results

```python
smell_word_counts = df['Smell_Word'].value_counts()
smell_word_counts.head(30).plot(kind='bar', title='Most frequent smell words')
```

```python
smell_source_counts = df['Smell_Source'].value_counts()
smell_source_counts.head(30).plot(kind='bar', title='Most frequent sources of smells')
```

```python
quality_counts = df['Quality'].value_counts()
quality_counts.head(30).plot(kind='bar', title='Most frequent smell qualities')
```

```python
carrier_counts = df['Odour_Carrier'].value_counts()
carrier_counts.head(30).plot(kind='bar', title='Most frequent odour carriers')
```

```python
odorant_counts = df['Evoked_Odorant'].value_counts()
odorant_counts.head(30).plot(kind='bar', title='Most frequent evoked odorants')
```

```python
location_counts = df['Location'].value_counts()
location_counts.head(30).plot(kind='bar', title='Most frequent locations')
```

```python
perceiver_counts = df['Perceiver'].value_counts()
perceiver_counts.head(30).plot(kind='bar', title='Most frequent perceivers')
```

```python
time_counts = df['Time'].value_counts()
time_counts.head(30).plot(kind='bar', title='Most frequent times')
```

```python
circumstances_counts = df['Circumstances'].value_counts()
circumstances_counts.head(30).plot(kind='bar', title='Most frequent circumstances')
```
